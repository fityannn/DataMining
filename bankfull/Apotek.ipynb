```python
import pandas as pd

data_source = 'ProjectDataMining.csv'

df = pd.read_csv(data_source, delimiter=";")

df
```

```python
df["GrandTotal"].value_counts()
```

```python
df.isna().sum()
```

```python
# Mengubah nilai NaN menjadi 0 pada kolom 'Jan', 'Feb', dan 'Mar'
df[['Jan', 'Feb', 'Mar']] = df[['Jan', 'Feb', 'Mar']].fillna(0)

# Mengubah tipe data menjadi integer untuk kolom 'Jan', 'Feb', dan 'Mar'
df[['Jan', 'Feb', 'Mar']] = df[['Jan', 'Feb', 'Mar']].astype(int)

print(df)
```

```python
# DATA VISUALIZATION

import matplotlib.pyplot as plt

df[['Jan', 'Feb', 'Mar']].plot()
plt.title('Penjualan Bulan Jan, Feb, Dan Mar')
plt.xlabel('Index')
plt.ylabel('Penjualan')
plt.show()
```

```python
print(df[['Jan', 'Feb', 'Mar', 'GrandTotal']].describe().astype(int))
```

```python
correlation_matrix = df[['Jan', 'Feb', 'Mar', 'GrandTotal']].corr()
print(correlation_matrix)
```
