```python
'''Nama = Fityan Abdul Khalik
NIM = 10521035
Kelas = IS-1'''
import pandas as pd #mengimport library pandas dan mengganti namanya menjadi pd
import numpy as np #mengimport library numpy dan mengganti namanya menjadi py
import nltk #mengimport library pandas nltk
import string #mengimport library pandas string
from sklearn.model_selection import train_test_split #mengambil dan mengimport train_test_split dari library sklearn.model_selection
from sklearn.naive_bayes import MultinomialNB #mengambil dan mengimport MultinomialNB dari library sklearn.naive_bayes
from sklearn.neighbors import KNeighborsClassifier #mengambil dan mengimport KNeighborsClassifier dari library sklearn.neighbors
from sklearn.tree import DecisionTreeClassifier #mengambil dan mengimport DecisionTreeClassifier dari library sklearn.tree
from sklearn.ensemble import RandomForestClassifier #mengambil dan mengimport RandomForestsClassifier dari library sklearn.ensemble
from sklearn.linear_model import LogisticRegression #mengambil dan mengimport LogisticRegression dari library sklearn.linear_model
from sklearn.svm import SVC #mengambil dan mengimport SVC dari library sklearn.svm
from sklearn.neural_network import MLPClassifier #mengambil dan mengimport MLPClassifier dari library sklearn.neural_network
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix #mengambil dan mengimport classification_report , accuracy_score , confusion_matrix dari library sklearn.metrics
import seaborn as sns #mengimport library seaborn dan mengganti namanya menjadi sns
import matplotlib.pyplot as plt #mengimport library matplotlib.pyplot dan mengganti namanya menjadi plt

X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.3, random_state = 1)

X_train.shape

NaiveBayes = MultinomialNB().fit(X_train,np.ravel(y_train,order='C'))
Knn = KNeighborsClassifier(n_neighbors = 3).fit(X_train,np.ravel(y_train,order='C'))
RandomForest = RandomForestClassifier(n_estimators=50, max_depth=3).fit(X_train,np.ravel(y_train,order='C'))
DTree =    DecisionTreeClassifier().fit(X_train,np.ravel(y_train,order='C'))
MultiLP =   MLPClassifier(max_iter= 100).fit(X_train,np.ravel(y_train,order='C'))
SuppVM =  SVC(gamma='scale', decision_function_shape='ovo', kernel = 'linear').fit(X_train,np.ravel(y_train,order='C'))

models = [
    NaiveBayes,
    Knn,
    RandomForest, 
    DTree, 
    MultiLP, 
    SuppVM, 
] 
dlist = [
    'NB',
    'K-NN',
    'RF', 
    'DT', 
    'MLP', 
    'SVM', 
]

i = 0
print(y_test.shape)
entries = []
for model in models:
  prediction = model.predict(X_test)
  accuracies = accuracy_score(y_test, prediction)
  nameS = dlist[i]
  entries.append((nameS, model, accuracies))
  i = i + 1

cv_df = pd.DataFrame(entries, columns=['Classifier','prediction','accuracy'])
cv_df.to_csv('akurasi.csv')
print(cv_df)


sns.boxplot(x='Classifier', y='accuracy', data=cv_df)
sns.stripplot(x='Classifier', y='accuracy', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=2)

print(cv_df.groupby('accuracy').accuracy.mean())
plt.show()
```
