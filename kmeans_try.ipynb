```python
import matplotlib.pyplot as plt

x = [410,	360,	346,	255,	292,	266,	345,	387,	408,	273,	81,	223,	110,	116]
y = [44,	149,	334,	161,	288,	141,	261,	489,	366,	141,	475,	365,	82,	59]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score, adjusted_rand_score, adjusted_mutual_info_score, silhouette_score
 

print(f"Homogeneity: {homogeneity_score(x, y):.3f}")
print(f"Completeness: {completeness_score(x, y):.3f}")
print(f"V-measure: {v_measure_score(x, y):.3f}")
print(f"Adjusted Rand Index: {adjusted_rand_score(x, y):.3f}")
print(
    "Adjusted Mutual Information:"
    f" {adjusted_mutual_info_score(x, y):.3f}"
)
```

```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

iris = datasets.load_iris()

x = iris.data[:,0]       #x = df['']   y = df['']
y = iris.data[:,1]

print(x)
print(y)

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
